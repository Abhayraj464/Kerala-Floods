```python
x = new.drop(columns=['FLOODS'])
```

```python
y = new['FLOODS']
y
```

### Model creation 

#### Fit the scaler to the data and transform the data

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled.shape
```

#### split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,train_size= 0.3, random_state=42)
```

#### Train the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
fd = LogisticRegression()
```

```python
fd.fit(x_train,y_train)
```

#### Predict the target variable on the test data

```python
y_pred = fd.predict(x_test)
y_pred.shape
```

```python
y_test.shape
```

#### Evaluating the model

```python
from sklearn.metrics import accuracy_score, classification_report,precision_score,recall_score,f1_score,confusion_matrix
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy : {accuracy: .2f}")
```

```python
classification = classification_report(y_test,y_pred)
print(classification)
```

```python
matrix = confusion_matrix(y_test,y_pred)
matrix
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(matrix, annot =True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

- Classification Report:

    - Class 0 (No Floods):
        - Precision: 97% (out of all predicted "No Floods," 97% were correct)
        - Recall: 83% (out of all actual "No Floods," 83% were correctly identified)
        - F1 Score: 90%
    - Class 1 (Floods):
        - Precision: 85% (out of all predicted "Floods," 85% were correct)
        - Recall: 98% (out of all actual "Floods," 98% were correctly identified)
        - F1 Score: 91%
        
        
<b>Overall:
The model performs well across both classes, with a balanced precision, recall, and F1 score.
Confusion Matrix:


- The confusion matrix shows:
    - True Positives (TP): 40 instances of Class 1 (Floods) were correctly identified.
    - True Negatives (TN): 35 instances of Class 0 (No Floods) were correctly identified.
    - False Positives (FP): 7 instances of Class 0 (No Floods) were incorrectly classified as Floods.
    - False Negatives (FN): 1 instance of Class 1 (Floods) was incorrectly classified as No Floods.
